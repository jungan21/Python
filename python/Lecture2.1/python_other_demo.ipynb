## Sort

```python
nums = [1, 2, 5, 4, 3]
```

```python
sorted_nums = sorted(nums)    # Sorted排序生成新的list, 原来的list不变
sorted_nums
```

```python
nums
```

```python
nums.sort(reverse=True)    # Sort原地排序
nums
```

### 参数key可以指定排序的依据

```python
tuples = [(1, 2), (2, -3), (-10, 5), (-13, 0)]
```

```python
sorted(tuples, key=lambda x: x[1])
```

```python
sorted(tuples, key=lambda x: x[0] + x[1])
```

```python
from functools import cmp_to_key
def cmp(a, b):
    return (a[0] + a[1] - b[0] - b[1])
sorted(tuples, key=cmp_to_key(cmp))
```

## Lambda表达式

```python
func = lambda x : x ** 2
func(5)
```

```python
func2 = lambda x, y=2 : x + y
func2(4)
```

## Map函数

```python
nums = [1, 2, 3, 4]
def square(x):
    return x * x
list(map(square, nums))
```

```python
list(map(lambda x : x * x, nums))
```

## Filter函数

```python
fit = filter(lambda x : x % 2 == 0, [1, 2, 3, 4, 5])
for num in fit:
    print(num)
```

## Reduce函数

```python
from functools import reduce
reduce(lambda x, y: x * y, [1, 2, 3, 4])
```
